```python
from google.colab import drive
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
# Getting CSV Files
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/predictions_noc_stock.csv')
start_date = datetime(2024,1,30)
# Check the data type of "Unnamed: 0" column
if df["Unnamed: 0"].dtype != np.datetime64:
    try:
        df["Unnamed: 0"] = pd.to_datetime(df["Unnamed: 0"])
    except:
        raise TypeError("Column 'Unnamed: 0' cannot be converted to datetime format.")

# Filter the DataFrame
df_filtered = df[df["Unnamed: 0"] >= start_date]

# Print the filtered DataFrame
print(df_filtered)
plt.figure(figsize=(12, 6))
plt.plot(df_filtered["Unnamed: 0"], df_filtered["Predicted_Close"])



plt.xlabel("Date")
plt.ylabel("Predicted Close Price")

plt.tight_layout()
plt.show()
```
